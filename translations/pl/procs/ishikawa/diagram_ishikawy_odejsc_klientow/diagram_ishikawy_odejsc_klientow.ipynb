## Streszczenie

Ta analiza porządkuje przyczyny źródłowe podwyższonych odejść klientów SaaS w ustrukturyzowany diagram przyczynowo-skutkowy (Ishikawy). Segment B2B mid-market odchodzi w tempie około 18% rocznie wobec celu 8%. Dwadzieścia ustaleń dotyczących przyczyn źródłowych — zebranych z ankiet odejścia, wywiadów zespołu Customer Success oraz atrybucji kohortowej — zgrupowano w sześć kategorii przyczyn i przedstawiono za pomocą `PROC FISHBONE` jako diagram złożony z 27 węzłów (jeden skutek, sześć głównych „ości” kategorii, dwadzieścia podprzyczyn).

Towarzyszące zestawienie `PROC MEANS` pokazuje, gdzie koncentruje się przypisywalny wskaźnik odejść. **Produkt** ma najwięcej ustaleń (5 przyczyn, 48,1 łącznych punktów wpływu), a za nim **Cennik** (4 przyczyny, 42,3) i **Wsparcie** (4 przyczyny, 40,2). Mierząc średnim wpływem *na przyczynę*, najbardziej dotkliwą kategorią jest **Konkurencja** (średnia 11,8), a pojedynczym ustaleniem o największym wpływie ogółem jest podwyżka ceny odnowienia w kategorii **Cennik** (14,2). Widok Pareto z `PROC SGPLOT` porządkuje każdą pojedynczą przyczynę, dzięki czemu kierownictwo widzi, którymi „ośćmi” zająć się najpierw. Ten materiał zasila kwartalny przegląd retencji zespołu Customer Success.

| Źródło | Opis | Rekordy |

```jenner
/* --------------------------------------------------------
   Ustalenia dotyczące przyczyn źródłowych odejść klientów (RCA)
   Platforma SaaS, segment B2B mid-market.
   Każdy wiersz: kategoria przyczyny, konkretna przyczyna,
   notatka pomocnicza z dowodem/podprzyczyną oraz szacowany
   udział odejść jej przypisywany (z ankiet odejścia
   i atrybucji kohortowej).
   -------------------------------------------------------- */
DANE work.churn_rca;
    INFILE DATALINES dlm='|' dsd truncover;
    WEJŚCIE category :$15. cause :$90. subcause :$110. churn_impact_pct;
    ETYKIETA category="Kategoria" cause="Przyczyna" subcause="Podprzyczyna"
        churn_impact_pct="Udział w odejściach (%)";
    DATALINES;
Produkt|Luki w kluczowych funkcjach względem konkurentów klasy enterprise|Eksport masowy i limity zapytań API wskazywane w 34% ankiet odejścia|12.4
Produkt|Ograniczone możliwości raportowania|Brak kreatora niestandardowych pulpitów w planie standardowym|9.8
Produkt|Aplikacja mobilna w tyle za funkcjonalnością wersji webowej|3 kluczowe procesy niemożliwe do ukończenia przez użytkowników tylko mobilnych|8.2
Produkt|Częste zmiany interfejsu bez uprzedzenia|Zaawansowani użytkownicy zakłóceni kwartalnymi przeprojektowaniami|7.6
Produkt|Luki w rynku integracji|Brak natywnych konektorów do Salesforce i HubSpot|10.1
Cennik|Podwyżka ceny odnowienia bez wcześniejszego ostrzeżenia|Wzrost MRR o 18% z 30-dniowym wyprzedzeniem|14.2
Cennik|Model per-seat karze rozwój zespołu|Koszt podwaja się przy wzroście zespołu z 10 do 20 osób|11.3
Cennik|Rabat początkowy nieodnawialny|Szok cenowy przy pierwszym odnowieniu|9.7
Cennik|Brak zachęty do umowy wieloletniej|Cena roczna identyczna jak miesięczna|7.1
Wsparcie|Czas pierwszej odpowiedzi przekracza 24 godziny|Benchmark dla tego planu to 4 godziny|10.8
Wsparcie|Luka jakości wsparcia w weekendy|Zespół offshore bez znajomości produktu|9.3
Wsparcie|Brak dedykowanego CSM poniżej planu Enterprise|Klienci mid-market bez przypisanego opiekuna|11.5
Wsparcie|Niejasna ścieżka eskalacji dla klientów|38% odchodzących klientów wskazało zamieszanie w eskalacji|8.6
Wdrożenie|Czas do uzyskania wartości przekracza 90 dni|Złożoność wdrożenia nieujawniona przed sprzedażą|12.9
Wdrożenie|Nieaktualna dokumentacja samoobsługowa|Ostatnia aktualizacja 11 miesięcy temu|8.1
Wdrożenie|Brak przewodników w aplikacji dla nowych funkcji|Adopcja funkcji poniżej 30% po 6 miesiącach|9.4
Konkurencja|Konkurent oferuje cenę niższą o 30%|Trzej kluczowi klienci wskazali konkretnego konkurenta|13.1
Konkurencja|Konkurent oferuje cennik bez limitu miejsc|Eliminuje karę per-seat dla rosnących zespołów|10.6
Relacje|Niezarządzana rotacja sponsora wykonawczego|Odejście lidera projektu; brak programu ponownego zaangażowania|11.7
Relacje|Częstotliwość QBR zredukowana do rocznej|Redukcja zasobów zespołu CS 18 miesięcy temu|9.2
;
WYKONAJ;
```

---

```jenner
/* --------------------------------------------------------
   Punkt odniesienia: koncentracja wpływu na odejścia według kategorii RCA.
   Suma pokazuje, gdzie koncentrują się przypisywalne odejścia; max
   wskazuje pojedynczą najgorszą przyczynę w każdej kategorii.
   -------------------------------------------------------- */
PROCEDURA ŚREDNIE DANE=work.churn_rca n sum mean MAX maxdec=1;
    KLASA category;
    ZMIENNA churn_impact_pct;
    TYTUŁ 'Analiza RCA odejść klientów: udział procentowy według kategorii';
WYKONAJ;
```

---

```jenner
/* --------------------------------------------------------
   Widok Pareto: pojedyncze przyczyny źródłowe uszeregowane według
   szacowanego wpływu na odejścia, kolorowane według kategorii.
   Uzupełnia to diagram Ishikawy, kwantyfikując, które "ości" mają
   największe znaczenie.
   -------------------------------------------------------- */
PROCEDURA SORTUJ DANE=work.churn_rca;
    WEDŁUG MALEJĄCO churn_impact_pct;
WYKONAJ;

PROCEDURA SGPLOT DANE=work.churn_rca;
    HBAR cause / RESPONSE=churn_impact_pct
        GROUP=category;
    XAXIS ETYKIETA='Szacowany wpływ na odejścia (%)';
    YAXIS ETYKIETA='Przyczyna źródłowa' discreteorder=data;
    keylegend / TYTUŁ='Kategoria przyczyny' position=bottom;
    TYTUŁ 'Przyczyny źródłowe odejść klientów uszeregowane wg szacowanego wpływu';
WYKONAJ;
```

### Diagram przyczynowo-skutkowy (Ishikawy)

`PROC FISHBONE` buduje diagram bezpośrednio z `WORK.CHURN_RCA`: każda odrębna wartość `category` staje się główną „ością” odchodzącą od centralnej osi, a poszczególne ustalenia `cause` wiszą na swojej „ości” kategorii jako podprzyczyny. Procedura drukuje poniżej pełną strukturę przyczynowo-skutkową — skutek na czele, sześć „ości” kategorii oraz każdą podprzyczynę pogrupowaną pod właściwą „ością” — i zapisuje odpowiadającą grafikę diagramu na dysku. Widok Pareto powyżej to towarzysząca perspektywa, która kwantyfikuje, która z tych „ości” niesie najwięcej przypisywalnych odejść.

```jenner
/* --------------------------------------------------------
   PROC FISHBONE: diagram przyczynowo-skutkowy odejść klientów.
   Diagram jest budowany bezpośrednio z WORK.CHURN_RCA, więc
   sześć kategorii przyczyn i ich przyczyny to dokładnie te
   ustalenia, które analizowano powyżej. CATEGORY tworzy główne
   "ości"; CAUSE wisi jako podprzyczyny na każdej "ości".
   -------------------------------------------------------- */
PROCEDURA fishbone DANE=work.churn_rca;
    cause category / subcause=cause;
    effect 'Podwyższone odejścia klientów (18% rocznie wobec celu 8%)';
    TYTUŁ 'Analiza przyczyn źródłowych odejść klientów - segment mid-market';
    footnote 'Źródła: ankiety odejścia, wywiady zespołu CS, atrybucja kohortowa';
WYKONAJ;
```

### Interpretacja

Struktura przyczynowo-skutkowa oraz wskaźniki `PROC MEANS` w podziale na kategorie wskazują kierownictwu niewielką liczbę posunięć o dużej dźwigni. Dwa najwyżej punktowane pojedyncze ustalenia — podwyżka ceny odnowienia bez ostrzeżenia w kategorii **Cennik** (14,2) oraz ustalenie z kategorii **Konkurencja**, że rywal sprzedaje równoważne możliwości o 30% taniej (13,1) — mają charakter komercyjny, a nie produktowy, co sugeruje, że przejrzystość cenowa i pozycjonowanie konkurencyjne to najpilniejsze dźwignie krótkoterminowe. Przekroczenie 90 dni czasu do uzyskania wartości w kategorii **Wdrożenie** (12,9) to największe ustalenie po stronie usług, bezpośrednio powiązane z ustalaniem oczekiwań przed sprzedażą.

Pod względem liczby ustaleń **Produkt** dominuje w diagramie (5 z 20 ustaleń, 48,1 łącznych punktów wpływu), ale jego ustalenia są indywidualnie mniejsze (średnia 9,6) niż w kategoriach **Konkurencja** (średnia 11,8) czy **Cennik** (średnia 10,6) — przypomnienie, że długa lista luk produktowych nie jest automatycznie najwyższym priorytetem. „Ość” **Relacje** (nieutrzymywana rotacja sponsora wykonawczego i zredukowana częstotliwość QBR, łącznie 20,9) reprezentuje zmiany procesowe, które zespół CS może wdrożyć w bieżącym kwartale bez inwestycji produktowych czy cenowych, oferując najszybszą ścieżkę do wymiernej poprawy retencji. Pełna uszeregowana lista na wykresie Pareto pozwala kierownictwu sekwencjonować interwencje według szacowanego wpływu, a nie według wielkości kategorii.
