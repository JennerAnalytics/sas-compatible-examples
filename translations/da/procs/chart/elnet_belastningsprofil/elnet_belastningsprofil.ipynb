# Profilering af regionalt netbelastning og udfald med PROC CHART

PROC CHART producerer tegnbaserede søjle-, blok- og cirkeldiagrammer direkte i listingen — ingen ODS Graphics-enhed kræves. Det gør den til et hurtigt førstehåndsudforskningsværktøj for et netdriftsteam, der vil *se* formen på sine belastnings- og pålidelighedsdata, før det bygger polerede GCHART- eller SGPLOT-visualiseringer.

I denne notebook:

1. Genererer vi en syntetisk måned af målaflæsninger fra fordelingskredse for en regional netoperatør.
2. Diagrammerer **produktionsmikset** (andel af aflæsninger pr. kilde).
3. Sammenligner **gennemsnitlig og samlet leveret belastning** på tværs af forsyningsregioner.
4. Afdækker **aftenens efterspørgselstop** pr. time på dagen.
5. Bruger et **blokdiagram** til at krydse region mod produktionskilde.
6. Rangordner **udfaldsminutter** pr. kilde for at finde den mindst pålidelige forsyning.

Hver sætning og indstilling nedenfor er standard SAS 9.4 PROC CHART-syntaks.

## Trin 1 — Generér de syntetiske driftsdata

DATA-trinnet nedenfor fabrikerer målaflæsninger i en løkke på 1.200 iterationer. Hver aflæsning tildeles en forsyningsregion, en produktionskilde (Gas dominerer, med Vind, Sol og Vand som resten) og en time på dagen. Belastningen er højere i aftenvinduet 17:00–21:00 (og vi markerer disse aflæsninger som spidsbelastning), og vi trækker udfaldsminutter fra en Poisson-fordeling. `call streaminit` fastlægger det tilfældige startpunkt, så data er reproducerbare.

NOTE i loggen viser det praktiske resultat: denne kørsel er i ulicenseret tilstand, så det gemte `grid_ops`-datasæt er begrænset til de første 100 af disse aflæsninger (100 rækker, 7 kolonner). Hvert diagram, der følger, opsummerer den 100-rækkers stikprøve, og hver PROC CHART-log bekræfter, at den læste 100 rækker.

```jenner
/* Syntetiske fordelingskreds-operationer for en regional netoperatør */
data grid_ops;
    CALL streaminit(20260531);
    LÆNGDE region $12 source $9 peak_flag $3;
    GØR meter_id = 1 TIL 1200;
        r = ceil(4 * rand('uniform'));
        HVIS r = 1 SÅ region = 'Nord';
        ELLERS HVIS r = 2 SÅ region = 'Syd';
        ELLERS HVIS r = 3 SÅ region = 'Øst';
        ELLERS region = 'Vest';
        u = rand('uniform');
        HVIS u < 0.42 SÅ source = 'Gas';
        ELLERS HVIS u < 0.70 SÅ source = 'Vind';
        ELLERS HVIS u < 0.88 SÅ source = 'Sol';
        ELLERS source = 'Vand';
        hour = floor(24 * rand('uniform'));
        BASE = 18 + 5 * (region = 'Nord') + 3 * (region = 'Vest');
        HVIS hour >= 17 AND hour <= 21 SÅ GØR;
            load_mwh = BASE + 12 + 6 * rand('normal');
            peak_flag = 'Ja';
        SLUT;
        ELLERS GØR;
            load_mwh = BASE + 4 * rand('normal');
            peak_flag = 'Nej';
        SLUT;
        HVIS load_mwh < 0 SÅ load_mwh = 0;
        outage_min = rand('poisson', 2.5);
        UDDATA;
    SLUT;
    FJERN r u BASE;
KØR;
```

## Trin 2 — Produktionsmiks

Hvilken andel af aflæsningerne står hver produktionskilde for? Et lodret søjlediagram med `TYPE=PERCENT` besvarer dette direkte: søjlehøjderne er procentdelen af alle observationer, der falder i hver kildekategori. Fordi `source` er en tegnvariabel, behandler PROC CHART automatisk dens værdier som diskrete kategorier.

```jenner
PROCEDURE chart data=grid_ops;
    VBAR source / type=PERCENT;
    MÆRKAT source="Kilde";
KØR;
```

## Trin 3 — Gennemsnitlig leveret belastning pr. region

Nu skifter vi fra at tælle til at opsummere en måling. Ved at navngive `load_mwh` i `SUMVAR=` med `TYPE=MEAN` bliver søjlelængden den gennemsnitlige belastning pr. region, og vi anmoder om statistikkolonnerne eksplicit: `MEAN` udskriver gennemsnittet ved siden af hver søjle, og `CFREQ` tilføjer en kumuleret frekvenskolonne. Nord bærer den højeste gennemsnitlige leverede belastning (gennemsnit omkring 28,0 MWh), i overensstemmelse med den regionale forskydning indbygget i data, mens Syd er lavest (omkring 19,8 MWh).

Vi videregiver også `ASCENDING` for at ordne søjlerne fra laveste til højeste gennemsnit. Diagrammet viser regionerne i netop den rækkefølge — Syd, Øst, Vest, Nord med gennemsnit 19,8, 21,7, 24,2 og 28,0 MWh — så søjlerne løber fra korteste til længste. Uanset rækkefølgen kan den nøjagtige rangordning altid aflæses i den udskrevne `Mean`-kolonne.

```jenner
PROCEDURE chart data=grid_ops;
    HBAR region / SUMVAR=load_mwh type=mean
                  cfreq mean ascending;
    MÆRKAT region="Region" load_mwh="Belastning (MWh)";
KØR;
```

## Trin 4 — Samlet belastning pr. region

Her gør `TYPE=SUM` hver søjle til den *samlede* leverede belastning for regionen frem for gennemsnittet, så de højere søjler markerer de regioner, der leverer mest samlet energi på tværs af de stikprøvetagne aflæsninger. Nord fører igen på samlet leveret belastning.

Sætningen anmoder også om `SUBGROUP=peak_flag`, som beder PROC CHART om at opdele hver søjle efter, om aflæsningen faldt i aftenens spidsbelastningsvindue. Hver søjle segmenteres med en distinkt glyf pr. undergruppeniveau — `J` for spidsaflæsninger (`Ja`) og `N` for ikke-spids (`Nej`) — og en forklaring under diagrammet knytter symbolerne til `Spidsbelastning`-værdierne. Den øverste `J`-del af hver søjle viser, hvor stor en andel af den samlede belastning der lander i aftenens spidsvindue.

```jenner
PROCEDURE chart data=grid_ops;
    VBAR region / SUMVAR=load_mwh type=sum
                  SUBGROUP=peak_flag;
    MÆRKAT region="Region" load_mwh="Belastning (MWh)" peak_flag="Spidsbelastning";
KØR;
```

## Trin 5 — Belastningsformen henover dagen

`hour` er kontinuert, så vi fastlægger selv opdelingen med en eksplicit `MIDPOINTS=`-liste ved 4-timers centre (2, 6, 10, 14, 18, 22). Hver søjle viser den gennemsnitlige leverede belastning for aflæsninger nær den time. Søjlen centreret om 18 bør skille sig ud — det er aftenens efterspørgselstop, vi byggede ind i data.

```jenner
PROCEDURE chart data=grid_ops;
    VBAR hour / SUMVAR=load_mwh type=mean
                midpoints=2 6 10 14 18 22;
    MÆRKAT hour="Time på dagen" load_mwh="Belastning (MWh)";
KØR;
```

## Trin 6 — Region efter produktionskilde (blokdiagram)

Et BLOCK-diagram gengiver en lille tovejstabel som et gitter af blokke. Med `GROUP=source` og `SUMVAR=load_mwh / TYPE=MEAN` krydser diagrammet hver region mod den produktionskilde, der betjener den, med blokhøjde proportional med gennemsnitsbelastningen — en kompakt måde at få øje på, hvilke region/kilde-kombinationer der bærer den tungeste gennemsnitlige belastning.

```jenner
PROCEDURE chart data=grid_ops;
    BLOCK region / SUMVAR=load_mwh type=mean
                   GROUP=source;
    MÆRKAT region="Region" source="Kilde" load_mwh="Belastning (MWh)";
KØR;
```

## Trin 7 — Produktionsmiks som cirkeldiagram

Den samme kildeandelsinformation fra Trin 2, tegnet som en cirkel. PIE med `TYPE=PERCENT` dimensionerer hvert udsnit efter dets procentdel af de samlede aflæsninger og udskriver en forklaring med udsnitsprocenter ved siden af figuren.

```jenner
PROCEDURE chart data=grid_ops;
    PIE source / type=PERCENT;
    MÆRKAT source="Kilde";
KØR;
```

## Trin 8 — Udfaldsminutter pr. kilde

Til sidst rangordner vi pålideligheden. `SUMVAR=outage_min` med `TYPE=SUM` totaliserer udfaldsminutter pr. kilde. Vi videregiver `DESCENDING` for at forsøge at flyde den dårligst præsterende kilde til toppen, men i modsætning til `ASCENDING` i Trin 3 har `DESCENDING` endnu ingen effekt her — søjlerne udskrives i den almindelige kategorirækkefølge (Vand, Vind, Gas, Sol). Læs derfor rangordningen fra den udskrevne `Sum`-kolonne: Gas står for de fleste samlede udfaldsminutter i denne stikprøve (122), et godt stykke foran Vind (64), Sol (43) og Vand (38). Det følger produktionsmikset — Gas betjener de fleste aflæsninger, så den akkumulerer de fleste udfaldsminutter samlet set.

```jenner
PROCEDURE chart data=grid_ops;
    HBAR source / SUMVAR=outage_min type=sum descending;
    MÆRKAT source="Kilde" outage_min="Udfaldsminutter";
KØR;
```

## Fortolkning af resultaterne

At læse diagrammerne sammen giver driftsteamet et hurtigt situationsbillede (over den 100-rækkers stikprøve, som denne kørsel indfangede):

- **Produktionsmiks (Trin 2 og 7).** Gas bærer den største andel af aflæsninger (omkring 45 %), med Vind som nummer to (omkring 28 %) og Vand og Sol bagest (omkring 14 % og 13 %) — det lodrette søjlediagram og cirklen fortæller den samme historie på to måder, en nyttig fornuftskontrol.
- **Belastning pr. region (Trin 3 og 4).** Gennemsnitsbelastningens HBAR viser Nord med den højeste gennemsnitlige leverede belastning (gennemsnit ~28 MWh) og Syd den laveste (~20 MWh), i overensstemmelse med den regionale forskydning indbygget i data. SUM-diagrammet bekræfter, at Nord også leverer mest samlet energi.
- **Daglig belastningsform (Trin 5).** Midtpunktssøjlen centreret om time 18 er tydeligt den højeste, hvilket bekræfter efterspørgselstoppet 17:00–21:00, vi byggede ind i data — præcis dér, hvor et forsyningsselskab ville fokusere efterspørgselsstyring og kapacitetsplanlægning.
- **Pålidelighed (Trin 8).** Totalisering af udfaldsminutter pr. kilde afdækker Gas som den største bidragyder til nedetid i denne stikprøve (122 minutter), det naturlige næste mål for vedligeholdelsesgennemgang — selvom det for det meste afspejler, at Gas betjener de fleste aflæsninger.

Af de visningsindstillinger, der er brugt her, ordner `ASCENDING` nu søjlerne fra lavest til højest (Trin 3), og `SUBGROUP=` segmenterer hver søjle med en forklaring (Trin 4); `DESCENDING` (Trin 8) accepteres derimod, men har endnu ingen effekt, så den rangordning aflæses fra den udskrevne `Sum`-kolonne frem for fra søjlerækkefølgen.

PROC CHART er kun tegnbaseret output, så for bestyrelsesklare visualiseringer ville teamet flytte de samme visninger til PROC GCHART eller PROC SGPLOT. Men som et første gennemløb uden opsætning over et frisk udtræk besvarer disse diagrammer de operationelle spørgsmål — miks, størrelsesorden og timing — på sekunder.
